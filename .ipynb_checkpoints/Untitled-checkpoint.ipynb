```python
Emojis_Dict = [{'name':'happy', 'file_name':'1F642.png'},
               {'name':'sad', 'file_name':'1F630.png'},
               {'name':'cool', 'file_name':'1F60E.png'},
              ]
```

```python
#plt.figure(figsize=(9,9))
for i, d in enumerate(Emojis_Dict):
    #print(d['name'] + d['file_name'])
    plt.subplot(3,1,i+1)
    plt.imshow(plt.imread(d['file_name']))
    plt.xlabel(d['name'])
    plt.xticks([])
    plt.yticks([])
    plt.show()
```

```python
def create_example():
    class_id = np.random.randint(0, 3)
    image = np.ones((200, 200, 4))*255
    np1 = np.array(image)
    emojis = plt.imread(Emojis_Dict[class_id]['file_name'])
    #emojis = np.array(emojis)
    row = np.random.randint(0, 128)
    col = np.random.randint(0, 128)
    np1[row:row+72, col:col+72, :] = emojis
    return np1, class_id, (row+10), (col+10)
```

```python
image, class_id, row, col = create_example()
print(row, col)
```
